```python
# Aquí hemos importado las librerias necesarias para la ejecución de este programa
import matplotlib.pyplot as plt
import math
import networkx as nx
import numpy as np
```

En el notebook debes utilizar la función creada en el script `assignment`

```python
# Aquí importamos las funciones y datos del archivo assigment.py
from assignment import random_network
```

```python
# Aquí definimos las caracteristicas del Experimento 1
N1, P1 = 100, 0.01

# Aquí ejecutamos la función random_network
G1 = random_network(N1, P1)

# Aquí mostramos la matriz de adyacencia a traves de una figura
adj_matrix1 = nx.to_numpy_array(G1)
plt.figure(figsize=(6, 6))
plt.matshow(adj_matrix1, fignum=1, cmap='Greys')
plt.title('Matriz de Adyacencia (N=100, P=0.01)\n')
plt.colorbar()
plt.show()

# Aquí obtenemos el numero de enlaces reales
L1 = G1.number_of_edges()

# Aquí obtenemos el numero experado de enlaces
expected_L1 = P1 * math.comb(N1, 2)

# Aquí mostramos el numero de enlaces reales y el numero de enlaces esperados
print(f"Enlaces reales: {L1} | Enlaces esperados: {expected_L1}")

# Aquí obtenemos el grado medio
avg_degree1 = sum(dict(G1.degree()).values()) / N1
print(f"Grado medio: {avg_degree1}")

# Aquí mostramos un grafico circular
plt.figure(figsize=(6, 6))
nx.draw_circular(G1, node_size=30, node_color='skyblue', edge_color='gray')
plt.title('Red Aleatoria Circular')
plt.show()
```

```python
# Aquí definimos las caracteristicas del Experimento 2
N2, P2 = 100, 0.05

# Aquí ejecutamos la función random_network
G2 = random_network(N2, P2)

# Aquí obtenemos el numero de enlaces reales
L2 = G2.number_of_edges()

# Aquí obtenemos el numero de enlaces esperado
expected_L2 = P2 * math.comb(N2, 2)

# Aquí obtenemos el grado medio
avg_degree2 = sum(dict(G2.degree()).values()) / N2

# Aquí mostramos el numero de enlaces reales y el numero de enlaces esperado
print(f"Enlaces reales: {L2} | Esperados: {expected_L2}")

# Aquí mostramos el grado medio
print(f"Grado medio: {avg_degree2}")

# Aquí definimos las variables U, V
U, V = 0, 99

# Aquí intentamos obtener el camino mas corto
try:
    shortest_path = nx.shortest_path(G2, source=U, target=V)
    path_edges = list(zip(shortest_path, shortest_path[1:]))
    print(f"Camino más corto entre {U} y {V}: {shortest_path}")
except nx.NetworkXNoPath:
    shortest_path, path_edges = [], []
    print(f"No existe un camino entre {U} y {V}.")

# Aquí generamos el grafico
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(G2)
nx.draw(G2, pos, node_size=30, node_color='lightgray', edge_color='lightgray')

if shortest_path:
    nx.draw_networkx_nodes(G2, pos, nodelist=shortest_path, node_color='red', node_size=50)
    nx.draw_networkx_edges(G2, pos, edgelist=path_edges, edge_color='red', width=2)

plt.title('Red y Camino Más Corto (N=100, P=0.05)')
plt.show()
```
